```
import numpy as np
import random

weights = np.random.rand(X.shape[1])


def gradientDescent(X, y, weights, alpha, n, iteration):
    global losses
    losses = []
    for iteration_num in range(iteration):
        pred = X @ weights
        error = (pred - y)
        losses.append(((pred - y).T) @ ((pred - y) / (2 * n)))
        derivative = X.T @ error
        weights = weights - (alpha * derivative)
        print(f'Итерация {iteration_num} Потери: {losses[iteration_num]} Веса: {weights}')
    return weights, losses

```
